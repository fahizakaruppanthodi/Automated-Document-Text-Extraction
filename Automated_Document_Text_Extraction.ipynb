```python
#step 2
#import libraries
import pytesseract
from PIL import Image
from google.colab import files
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
#step 3
#define a function to perform OCR , certain image texts all read
def perform_ocr(image_path):
  try:
    img=Image.open(image_path)
    text=pytesseract.image_to_string(img)
    return text
  except Exception as e:
    return f"Error during OCR: {e}"
```

```python
#step 4 upload the image containing text
print("upload an image file containing text")
uploaded=files.upload()

if uploaded:
  uploaded_file_name=list(uploaded.keys())[0]
  print(f"uploaded image :{uploaded_file_name}")

  #optional display the uploaded image
  img=cv2.imread(uploaded_file_name)
  plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
  plt.title("uploded image")
  plt.axis("off")
  plt.show()

  extracted_text=perform_ocr(uploaded_file_name)
  print("\n--Extracted text-- ")
  print(extracted_text)
else:
  print("not image file uploaded")



```
